```python
from haystack.nodes import FARMReader
reader = FARMReader(model_name_or_path="distilbert-base-uncased-distilled-squad", use_gpu=True)
data_dir='data/'
```

```python
reader.train(data_dir=data_dir, train_filename="answers.json", use_gpu=True, n_epochs=1, save_dir="my_model")
```

```python
new_reader = FARMReader(model_name_or_path="my_model")
```

```python
context="Elon Musk believes that EVs are the future of transportation, and advancements in battery technology and charging infrastructure are key to accelerating their adoption.Electric vehicles (EVs) are vehicles that are powered by electric motors and use electricity stored in batteries or other energy storage devices for propulsion. Unlike conventional internal combustion engine (ICE) vehicles that rely on fossil fuels, EVs produce zero tailpipe emissions, making them more environmentally friendly and contributing to efforts in reducing greenhouse gas emissions and air pollution."
```

```python
from haystack import Pipeline, Document
from haystack.utils import print_answers
# reader = FARMReader(model_name_or_path="deepset/roberta-base-squad2")
p = Pipeline()
p.add_node(component=new_reader, name="Reader", inputs=["Query"])
res = p.run(
    query="what is the expert opinion? ", documents=[Document(content=context)]
)
print_answers(res,details="medium")
```
